```python
import requests
from bs4 import BeautifulSoup
import csv

url = 'http://www.saidarshan.org/baba/docs/saib.html'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

bajan_blocks = soup.find_all('pre')
meaning_blocks = soup.find_all('i')
print(len(bajan_blocks))
print(len(meaning_blocks))

data = []
for bajan_block, meaning_block in zip(bajan_blocks, meaning_blocks):
    bhajan = bajan_block.get_text().strip()
    meaning = meaning_block.get_text().strip().lstrip('(').rstrip(')')
    data.append(['Krishna', bhajan, meaning])

csv_file = 'baba_bhajans.csv'
with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Baba', 'Bhajan', 'Meaning'])
    writer.writerows(data)

print(f'Data successfully written to {csv_file}')
```
